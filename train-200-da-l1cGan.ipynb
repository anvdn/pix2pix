# Training notebook

The goal of this notebook is to obtain pictures generated by the Pix2Pix model at different stages of training.

### 0. Libraries and dataset

```python
import utils
import model
```

```python
# set dataset name
dataset_name = 'facades'
```

## I. No transformation, batch size of 1

### 1. Instantiate dataloaders

```python
# instantiate datasets and dataloaders
datasets = {mode: utils.ImageDataset(dataset_name = dataset_name, transform=utils.data_transforms[mode], mode = mode) for mode in ['train', 'val']}
dataloaders = {'train': utils.DataLoader(datasets['train'], batch_size=1, shuffle=True),
                    'val': utils.DataLoader(datasets['val'], batch_size=1, shuffle=False)}
```

### 2. Create model

```python
# instantiate model
p2p = model.Pix2Pix().to(utils.device)
```

### 3. Train

```python
history = utils.train(model = p2p,
            n_epochs = 200, 
            display_step = 20, 
            save_step = 20,
            dataloaders = dataloaders, 
            filename = 'facades_long',
            lr = 2e-4, 
            lbd = 200, 
            loss_l1_true = True,
            loss_l2_true = False,
            loss_cGAN_true = True) 
```

## II. Plot image side by side

```python
import utils
import model
import torch
```

```python
p2p = model.Pix2Pix().to(utils.device)
p2p.load_state_dict(torch.load(utils.weights_path + '/' + 'facades_long_ep' + str(199) + '.pkl'))

for i in range(100):
    input_val, real_val = datasets['val'].__getitem__(i)
    utils.generate_images(model = p2p, input = input_val[None, :], real = real_val[None, :])
```

```python
# create model
p2p = model.Pix2Pix().to(utils.device)

input_val, real_val = datasets['train'].__getitem__(22)

for i in range(19,199,20):
    # load state dict
    p2p.load_state_dict(torch.load(utils.weights_path + '/' + 'facades_long_ep' + str(i) + '.pkl'))
    utils.generate_images(model = p2p, input = input_val[None, :], real = real_val[None, :])
    print(torch.nn.L1Loss()(real_val[None, :].to(utils.device), p2p.generator(input_val[None, :].to(utils.device))).item())
```
